#### Rellenando Datos Dulos por type 1

```python
df_3_Fire = df_3[df_3["Type 1"] == "Fire"].fillna("Fighting")
df_3_Bug = df_3[df_3["Type 1"] == "Bug"].fillna("Flying")
df_3_Dark = df_3[df_3["Type 1"] == "Dark"].fillna("Flying")
df_3_Dragon = df_3[df_3["Type 1"] == "Dragon"].fillna("Flying")
df_3_Electric = df_3[df_3["Type 1"] == "Electric"].fillna("Flying")
df_3_Fairy = df_3[df_3["Type 1"] == "Fairy"].fillna("Flying")
```

```python
df_3_Fighting = df_3[df_3["Type 1"] == "Fighting"].fillna("Psychic")
df_3_Flying = df_3[df_3["Type 1"] == "Flying"].fillna("Dragon")
df_3_Ghost = df_3[df_3["Type 1"] == "Ghost"].fillna("Grass")
df_3_Grass = df_3[df_3["Type 1"] == "Grass"].fillna("Poison")
df_3_Ground = df_3[df_3["Type 1"] == "Ground"].fillna("Flying")
```

```python
df_3_Ice = df_3[df_3["Type 1"] == "Ice"].fillna("Ground")
df_3_Normal = df_3[df_3["Type 1"] == "Normal"].fillna("Flying")
df_3_Poison = df_3[df_3["Type 1"] == "Poison"].fillna("Flying")
df_3_Psychic = df_3[df_3["Type 1"] == "Psychic"].fillna("Fairy")
df_3_Rock = df_3[df_3["Type 1"] == "Rock"].fillna("Ground")
df_3_Steel = df_3[df_3["Type 1"] == "Steel"].fillna("Psychic")
df_3_Water = df_3[df_3["Type 1"] == "Water"].fillna("Ground")
```

##### Concatenando

```python
df_concat = pd.concat([

    df_3_Fire,
    df_3_Grass,
    df_3_Bug,
    df_3_Dark,
    df_3_Dragon,
    df_3_Electric,
    df_3_Fairy,
    df_3_Fighting,
    df_3_Flying,
    df_3_Ghost,
    df_3_Ground,
    df_3_Ice,
    df_3_Normal,
    df_3_Poison,
    df_3_Psychic,
    df_3_Rock,
    df_3_Steel,
    df_3_Water,

])
```

```python
df_concat
```

```python
df_total = pd.concat([

    df_2_notnull,
    df_concat,

]).sort_index() #Metodo para ordenar los indices e menor a mayor
```
