```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_clean = pd.read_csv("df_clean.csv")
df_clean
```

```python
df_clean.info()
```

#### Agrupaciones categoricas de los datos

##### Visualizando cantidad de pokemones legendarios 

```python
df_1 = df_clean.copy()
df_1
```

```python
df_1_count = df_1["Legendary"].value_counts()
df_1_count
```

```python
plt.figure(figsize = (8,6))
plt.pie(df_1_count, labels = df_1_count.index, autopct='%1.1f%%')
plt.legend(df_1_count.index)
plt.show()
```

How many Pokemons exist with an Attack value greater than 150?

```python
df_2 = df_1.copy()
df_2
```

```python
# Visualizando pokemones cuyo valor en Attack es igual o mayor a 150
df_2[df_2["Attack"] >= 150].sort_values("Attack", ascending= False)
```

##### Select all pokemons with a Speed of 10 or less

```python
df_3 = df_2.copy()
df_3
```

```python
df_speed = df_3[df_3['Speed'] <= 10].sort_values("Speed", ascending = False)
df_speed
```

##### How many Pokemons have a Sp. Def value of 25 or less?

```python
df_Sp_Atk = df_3.loc[df_3['Sp. Atk'] <= 25].sort_values("Sp. Atk", ascending = False)
df_Sp_Atk
```

##### Select all the Legendary pokemons


```python
df_legendary = df_3[df_3["Legendary"] == True]
df_legendary
```

##### Find the outlier

```python
df_4 = df_3.copy()
df_4
```

##### How many Fire-Flying Pokemons are there?

```python
# Count of pokemons Type 1[Fire]
df_Fire = df_4[df_4['Type 1'] == "Fire"]
print(f"Hay {df_Fire.shape[0]} pokemones fuego tipo 1")
```

```python
#Count of pokemons Type 2[Flying]
df_Flying = df_4[df_4["Type 2"] == "Flying"]
print(f"Hay {df_Flying.shape[0]} pokemones Flying de tipo 2 ")
```
