##### Find the most powerful pokemon from the first 3 generations, of type water ?

```python
df_8 = df_7.copy()
df_8
```

```python
# Datos unicos en Generation
df_8['Generation'].unique()
```

```python
#Filtrarando datos con el metodo .isin
df_8_generation = df_8[df_8['Generation'].isin([1,2,3])]
df_8_generation
```

```python
df_max_gen = df_8_generation[df_8_generation['Type 1'].isin(['Water'])].sort_values('Total', ascending= False)
df_max_gen.head(9)
```

```python
df_max_grap = df_max_gen[df_max_gen['Total'] >= 635]
df_max_grap
```

```python
df_pivot = df_max_grap.pivot_table(index='Name', columns='Type 1', values='Total', aggfunc='sum')
df_pivot.plot(kind='bar', stacked=True, figsize=(10, 6))
plt.title('Stacked Bar Plot of Total Values by Type 1')
plt.xlabel('Name of Pokemon')
plt.ylabel('Total')
plt.legend(title='Type 1')
plt.xticks(rotation=30)
plt.show()
```
